### constant
* 변하지 않는 매개 변수

```python
# tf.constant 로 상수 선언해서 연산수행
x1 = tf.constant(5)
x2 = tf.constant(2)
x3 = tf.constant(3)

mul_result = tf.multiply(x1,x2)
add_result = tf.add(x2,x3)

sub_result = tf.subtract(mul_result,add_result)

print(f"add_result = {add_result}")
print(f"mul_result = {mul_result}")
print(f"sub_result = {sub_result}")
```

### tensorflow 로 Linear Regression model 생성

Result model: $ y = 2X -1$

```python
 ## data 선언
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]
```

```python

## 평균 0, 분산 1의 파라미터의 정규분포로 부터 값을 가져옴.
# 학습을 통해 업데이트가 되어 변화되는 모델의 파라미터인 w,b를 의미한다.
W=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
b=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
lr=tf.constant(0.001)

history = np.zeros((2000,3))
```

```python
import  matplotlib.pyplot as plt
errors = []
for i in range(2000):   # epoch
  total_error = 0
  for j in range(len(x_data)):  # batch
    WX =tf.matmul([x_data[j]], W)
    y_hat = tf.add(WX,b)

    error = tf.subtract(y_data[j], y_hat)  # True - predict

    diff_W = tf.multiply(x_data[j],error)
    diff_b = error

    diff_W = tf.multiply(diff_W,lr)
    diff_b = tf.multiply(diff_b,lr)   # lr*error

    W = tf.add(W,diff_W)   # W +  deltaW + lr * x  * error
    b = tf.add(b,diff_b)   # b + delta b + lr * x * error

    visual_error = tf.square(error)
    total_error += visual_error
  
  ## 모든 데이터에 따른 error 값
  history[i,:] = [(total_error/len(x_data))[0],W[0],b[0]]   # 강사님의 history 
  errors.append((total_error.numpy()[0][0]/len(x_data)))    # 나의 history
  if i % 100 ==0:
    print(f"epoch: {i}  Error : {total_error/len(x_data)} ")
  
plt.plot(errors)
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title(f'Train Error   lr = 0.001 ')
plt.show()
```

```python
## 학습끝난 후 W와 b로 예측 하기
print(f"W : {W}")
print(f"b : {b}")
print(f"input 3 : Predict{tf.add(tf.multiply(3, W), b)} Result = 5")
print(f"input 4 : Predict{tf.add(tf.multiply(4, W), b)} Result = 7")

print(f"input 5 : Predict{tf.add(tf.multiply(5, W), b)} Result = 9")
```

### Keras를 이용해서 Linear Regression 구현하기

```python
## data 선언
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]

test_data=[[9.]]
```

```python
## tf.keras를 활용한 perceptron 모델 구현.
model =tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim =1) ) # 선언된 모델에 add를 통해 쌓아감. , 현재는 입력 변수 갯수 1, perceptron 1개.
model.summary() ## 설계한 모델 프린트
```

```python
optimizer = tf.keras.optimizers.SGD(lr = 0.01)  # 경사하강법 
loss = tf.keras.losses.mse               # Mean Square Error
metrics = tf.keras.metrics.mae          # Mean Absolute Error
```

```python
model.compile(loss = loss, optimizer = optimizer, metrics = [metrics])

model.fit(x_data, y_data, epochs = 2000, batch_size = 4)
```
