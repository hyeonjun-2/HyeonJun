```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras import activations
from tensorflow.keras import layers

x = tf.convert_to_tensor([[0, 1]], dtype=tf.float32)
```

### 문제 1
tf.kreas.Model을 상속받아 MyNeuralNetwork 클래스를 만들고 다음의 신경망을 설계하시오.
- 4개의 은닉층과 1개의 출력층으로 구성
- 은닉층의 뉴런의 수는 100개, 출력층의 뉴런의 수는 2개
- 은닉층의 활성화함수는 시그모이드, 출력층의 활성화함수는 선형


```python
class MyNeuralNetwork(keras.Model):
  def __init__(self):
    super(MyNeuralNetwork,self).__init__()
    self.layers1 = layers.Dense(100,activation = 'sigmoid')
    self.layers2 = layers.Dense(50, activation = 'sigmoid')
    self.layers3 = layers.Dense(25, activation = 'sigmoid')
    self.layers4 = layers.Dense(10, activation = 'sigmoid')
    self.layers5 = layers.Dense(2, activation = 'linear')

  def call(self, x):
    x1 = self.layers1(x)
    x2 = self.layers2(x1)
    x3 = self.layers3(x2)
    x4 = self.layers4(x3)
    return self.layers5(x4)
  
  
```

```python
model = MyNeuralNetwork()
y  = model(x)
print(y)
```

### 문제 2
tf.kreas.Model을 상속받아 MySequentialNeuralNetwork 클래스를 만들고 다음의 신경망을 설계하고, plot_model을 통해 layers를 보이시오.
- 99개의 은닉층과 1개의 출력층으로 구성
- 은닉층의 뉴런의 수는 100개, 출력층의 뉴런의 수는 2개
- 은닉층의 활성화함수는 시그모이드, 출령층의 활성화함수는 선형

```python
class MySequentialNeuralNetwork(keras.Model):
    # 정답을 작성해 주세요.
    def __init__(self):
      super(MySequentialNeuralNetwork,self).__init__()
      self.dense_layers = keras.Sequential()
      for i in range(99):
        self.dense_layers.add(layers.Dense(100,activation = 'sigmoid'))
      self.dense_layers.add(layers.Dense(2,activation = 'linear'))

    def call(self,x):
      return self.dense_layers(x)
```

```python
model = MySequentialNeuralNetwork()
y  = model(x)
print(y)
```
