```python
cv2_utils.invalidate_state()

img_list = glob.glob('/images/*.jpg')

cv2_utils.namedWindow('scene', cv2.WINDOW_NORMAL)

indx = 0
while True:
  img = cv2.imread(img_list[indx])
  if not cv2_utils.imshow('scene', img):
    break
  
  key_code = cv2.waitKey(1000)
  if key_code == 27:
    break
  
  indx += 1
  if indx >= 5:
    indx = 0

cv2_utils.destroyAllWindows()
```

```python
cv2_utils.invalidate_state()

src = cv2.imread(basepath + '/puppy.bmp')
hat = cv2.imread(basepath + '/imgbin_hat.png', cv2.IMREAD_UNCHANGED)

print(hat.shape)

mask = hat[:, :, 3]
hat_rgb = hat[:, :, 0:3]

apply_ratio = lambda x, mask_shape: int(x * (mask_shape[1] / mask_shape[0]))

mask_crop = mask[450:1980, 240:2100]
hat_rgb_crop = hat_rgb[450:1980, 240:2100]

src_crop = src[10:10 + 235, 230:230 + apply_ratio(235, mask_crop.shape)]
src_crop_shape = src_crop.shape[0:2]
mask_crop = cv2.resize(mask_crop, (src_crop_shape[1], src_crop_shape[0]))
hat_crop = cv2.resize(hat_rgb_crop, (src_crop_shape[1], src_crop_shape[0]))

cv2.copyTo(hat_crop, mask_crop, src_crop)
cv2_utils.imshow('src', src)
```
